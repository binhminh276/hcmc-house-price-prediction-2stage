# 1. DEMO PHÂN CỤM


## 1.1 HIERACHICAL

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

df = pd.read_csv('processed_dataset_cleaned_new.csv', sep=';', decimal=',')


features_to_cluster = ['Diện tích', 'Khoảng giá']

print(f"Đang gom cụm dựa trên tổ hợp: {features_to_cluster}")
X = df[features_to_cluster]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

silhouette_scores = []
K_range = range(2, 11)

print("Đang tính toán Silhouette Score...")
for k in K_range:
    hc_temp = AgglomerativeClustering(n_clusters=k, metric='euclidean', linkage='ward')
    cluster_labels = hc_temp.fit_predict(X_scaled)

    score = silhouette_score(X_scaled, cluster_labels)
    silhouette_scores.append(score)
    print(f"K = {k:2d} | Silhouette Score: {score:.4f}")

plt.figure(figsize=(10, 6))
plt.plot(K_range, silhouette_scores, marker='o', linestyle='-', color='g')
plt.title(f'Biểu đồ Silhouette Score - Hierarchical ({", ".join(features_to_cluster)})')
plt.xlabel('Số lượng cụm (K)')
plt.ylabel('Silhouette Score')
plt.xticks(K_range)
plt.grid(True)
plt.tight_layout()
plt.show()
```

##1.2 KMEANS

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

df = pd.read_csv('processed_dataset_cleaned_new.csv', sep=';', decimal=',')

features_to_cluster = ['Diện tích', 'Khoảng giá']

print(f"Đang gom cụm bằng K-Means dựa trên tổ hợp: {features_to_cluster}")
X = df[features_to_cluster]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

silhouette_scores = []
K_range = range(2, 16)

print("Đang tính toán Silhouette Score...")
for k in K_range:
    kmeans_temp = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans_temp.fit_predict(X_scaled)

    score = silhouette_score(X_scaled, cluster_labels)
    silhouette_scores.append(score)
    print(f"K = {k:2d} | Silhouette Score: {score:.4f}")

plt.figure(figsize=(10, 6))
plt.plot(K_range, silhouette_scores, marker='o', linestyle='-', color='b')
plt.title(f'Biểu đồ Silhouette Score - K-Means ({", ".join(features_to_cluster)})')
plt.xlabel('Số lượng cụm (K)')
plt.ylabel('Silhouette Score')
plt.xticks(K_range)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 1.3 OPTICs

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import OPTICS
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

df = pd.read_csv('processed_dataset_cleaned_new.csv', sep=';', decimal=',')


features_to_cluster = ['Diện tích', 'Khoảng giá']

print(f"Đang gom cụm bằng OPTICS dựa trên tổ hợp: {features_to_cluster}")
X = df[features_to_cluster]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

silhouette_scores = []
# Khảo sát các giá trị min_samples khác nhau
min_samples_range = range(5, 50, 2)

print("Đang tính toán Silhouette Score...")
valid_min_samples = []

for m in min_samples_range:
    optics_temp = OPTICS(min_samples=m)
    cluster_labels = optics_temp.fit_predict(X_scaled)

    unique_labels = set(cluster_labels) - {-1}

    if len(unique_labels) > 1:
        mask = cluster_labels != -1
        score = silhouette_score(X_scaled[mask], cluster_labels[mask])
        silhouette_scores.append(score)
        valid_min_samples.append(m)
        print(f"min_samples = {m:2d} | Số cụm: {len(unique_labels)} | Silhouette Score: {score:.4f}")
    else:
        print(f"min_samples = {m:2d} | Chỉ tìm thấy {len(unique_labels)} cụm (Không đủ để tính Silhouette)")

if len(valid_min_samples) > 0:
    plt.figure(figsize=(10, 6))
    plt.plot(valid_min_samples, silhouette_scores, marker='o', linestyle='-', color='r')
    plt.title(f'Biểu đồ Silhouette Score - OPTICS ({", ".join(features_to_cluster)})')
    plt.xlabel('Số lượng mẫu tối thiểu (min_samples)')
    plt.ylabel('Silhouette Score (loại trừ nhiễu)')
    plt.xticks(valid_min_samples)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("Không thể vẽ biểu đồ vì không có giá trị min_samples nào tạo ra >= 2 cụm.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import OPTICS
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('processed_dataset_cleaned_new.csv', sep=';', decimal=',')

features_to_cluster = ['Diện tích', 'Khoảng giá']
X = df[features_to_cluster]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Đang chạy thuật toán OPTICS với min_samples = 49...")
optics_model = OPTICS(min_samples=49)
cluster_labels = optics_model.fit_predict(X_scaled)

df['Cluster'] = cluster_labels

# 4. Trực quan hóa (Scatter Plot)
plt.figure(figsize=(12, 8))

# Tách riêng dữ liệu outlier (Nhãn -1) và dữ liệu thuộc các cụm (Nhãn >= 0)
outliers = df[df['Cluster'] == -1]
clusters = df[df['Cluster'] != -1]

# Vẽ các điểm thuộc cụm (dùng palette màu đa dạng)
sns.scatterplot(
    data=clusters,
    x=features_to_cluster[0],
    y=features_to_cluster[1],
    hue='Cluster',
    palette='tab10', # Bảng màu phân biệt rõ các cụm
    s=60,
    alpha=0.8,
    edgecolor='k',
    legend='full'
)

# Vẽ đè các điểm outlier lên với màu đen/xám và marker khác biệt
plt.scatter(
    outliers[features_to_cluster[0]],
    outliers[features_to_cluster[1]],
    color='gray',
    marker='x',
    s=30,
    alpha=0.6,
    label='Outliers (-1)'
)

plt.title(f'Phân bố cụm OPTICS (min_samples=49) - Bao gồm cả Outliers', fontsize=14)
plt.xlabel(features_to_cluster[0], fontsize=12)
plt.ylabel(features_to_cluster[1], fontsize=12)

plt.legend(title='Cụm (Cluster)', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# In ra thống kê số lượng
print("\nThống kê số điểm dữ liệu trong từng cụm:")
print(df['Cluster'].value_counts().sort_index())
```

#PHÂN CỤM
Dùng Kmeans, k=3

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import joblib

df = pd.read_csv('processed_dataset_cleaned_new.csv', sep=';', decimal=',')
cols = ['Diện tích', 'Khoảng giá', 'Giá m2']
features_to_cluster = ['Diện tích', 'Khoảng giá']

X = df[features_to_cluster]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

optimal_k = 3
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)

df['Cluster'] = kmeans_final.fit_predict(X_scaled)

joblib.dump(kmeans_final, 'kmeans_model.pkl')
joblib.dump(scaler, 'scaler.pkl')
print("Đã lưu mô hình K-Means (kmeans_model.pkl) và bộ Scaler (scaler.pkl) thành công!")

output_file = 'kmeans_clustered_full.csv'

df.to_csv(output_file, index=False, sep=';', decimal=',', encoding='utf-8')
print(f"Đã hoàn tất và lưu toàn bộ bảng dữ liệu vào file: {output_file}")

print("\n=== TRUNG BÌNH CÁC CHỈ SỐ THEO CỤM ===")
print(df.groupby('Cluster')[cols].mean().round(2))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# 1. Đọc dữ liệu
df = pd.read_csv('kmeans_clustered_full.csv', sep=';', decimal=',')

# 2. Tính Điểm Tiện ích (Score) từ 3 cột
df['Score Tiện ích'] = df['Gần bệnh viện'] + df['Gần chợ'] + df['Gần trường học']

# 3. Chỉ lấy nhóm Score 0 và Score 3 để so sánh
df_filtered = df[df['Score Tiện ích'].isin([0, 3])].copy()

# Đổi tên nhãn cho dễ nhìn
df_filtered['Nhóm Tiện ích'] = df_filtered['Score Tiện ích'].map({
    0: 'Không tiện ích (Score 0)',
    3: 'Tiện ích toàn diện (Score 3)'
})
df_filtered['Phân khúc'] = df_filtered['Cluster'].map({
    1: 'Cụm 1 (Bình dân)',
    0: 'Cụm 0 (Cao cấp)',
    2: 'Cụm 2 (Siêu cao cấp)'
})

# 4. Tính toán tỷ lệ phần trăm (Crosstab)
crosstab_df = pd.crosstab(df_filtered['Nhóm Tiện ích'], df_filtered['Phân khúc'], normalize='index') * 100

# Sắp xếp lại thứ tự cột cho hợp lý
crosstab_df = crosstab_df[['Cụm 1 (Bình dân)', 'Cụm 0 (Cao cấp)', 'Cụm 2 (Siêu cao cấp)']]

# 5. Vẽ biểu đồ 100% Stacked Bar Chart
plt.figure(figsize=(9, 6))
ax = crosstab_df.plot(
    kind='bar', stacked=True,
    color=['#1f77b4', '#ff7f0e', '#d62728'], # Xanh, Cam, Đỏ
    figsize=(9, 6), edgecolor='white', width=0.5
)

# 6. Tinh chỉnh hiển thị
plt.title('Cấu trúc Phân khúc (Cluster) bên trong nhóm Không tiện ích vs Đầy đủ tiện ích', fontsize=14, pad=20)
plt.xlabel('Nhóm Tiện ích', fontsize=12)
plt.ylabel('Tỷ lệ phần trăm (%)', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Thuộc Phân khúc', bbox_to_anchor=(1.05, 1), loc='upper left')

# Thêm ghi chú phần trăm lên cột
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    if height > 1: # Chỉ ghi số nếu > 1% để tránh rối mắt
        ax.annotate(f'{height:.1f}%',
                    (x + width/2, y + height/2),
                    ha='center', va='center',
                    color='white', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.savefig('cau_truc_tien_ich_cluster.png', dpi=300)
# plt.show()
```
