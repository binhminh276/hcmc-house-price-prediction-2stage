```python
import pandas as pd
import numpy as np
```

```python
df['area_per_bedroom'] = df['Diện tích'] / df['Số phòng ngủ']
df['area_per_bathroom'] = df['Diện tích'] / df['Số phòng tắm, vệ sinh']
```

```python
if "ID" not in df.columns:
      df.insert(0, "ID", range(1, len(df) + 1))
      print(f"Đã đánh ID thành công cho {len(df)} bản ghi!")

else:
      print("Cột ID đã tồn tại trong file.")
```

```python
df = df.drop(df[df['Diện tích'] > 10000].index)
```

```python
df = df.drop(df[df['Khoảng giá'] < 0.5].index)
```

```python
print("-" * 45)
print(f"BÁO CÁO LÀM SẠCH")
print("-" * 45)
print(f"Tổng số bản ghi ban đầu: {len(df)}")

df['area_per_bedroom'] = df['Diện tích'] / df['Số phòng ngủ']
df['area_per_bathroom'] = df['Diện tích'] / df['Số phòng tắm, vệ sinh']


outliers = (
    (df['Diện tích'] < 10) |
    (df['area_per_bedroom'] < 4) |
    (df['area_per_bathroom'] < 4) |
    (df['area_per_bedroom'] > 200) |
    (df['area_per_bathroom'] > 200)
)

unusual_ratio = (df['Số phòng ngủ'] / df['Số phòng tắm, vệ sinh'] > 5) | \
                (df['Số phòng tắm, vệ sinh'] / df['Số phòng ngủ'] > 5)


df = df[~outliers | unusual_ratio].copy()
print(f"Số bản ghi lỗi mật độ:   {outliers.sum()}")
print(f"Số bản ghi sạch còn lại: {len(df)}")
print("\nCơ cấu tập dữ liệu theo quy mô:")
print("-" * 45)
```

```python
df.drop('area_per_bathroom', axis = 1, inplace=True)
df.drop('area_per_bedroom', axis = 1, inplace=True)
```

```python
df.to_csv(
    "dataset_cleaned_outliers.csv",
    index=False,
    encoding="utf-8-sig",
    sep=";",
    decimal=",",
    float_format="%.6g"
)
```
