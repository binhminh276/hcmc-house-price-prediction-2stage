```python
import pandas as pd
import re
import numpy as np
```

```python
df = pd.read_csv("dataset_cleaned_outliers.csv", decimal=",", sep=";", encoding="utf-8")
df
```

```python
df['Mặt tiền'] = df['Mô tả'].str.contains(r'mặt tiền', flags=re.IGNORECASE, na=False).astype(int)
df['Gần bệnh viện'] = df['Mô tả'].str.contains(r'bệnh viện|bv|y dược|nhà thương', flags=re.IGNORECASE, na=False).astype(int)
df['Gần chợ'] = df['Mô tả'].str.contains(r'chợ|siêu thị|bách hóa|vinmart|coopmart|aeon', flags=re.IGNORECASE, na=False).astype(int)
df['Gần trường học'] = df['Mô tả'].str.contains(r'trường học|đại học|cấp 1|cấp 2|cấp 3|mầm non|mẫu giáo', flags=re.IGNORECASE, na=False).astype(int)
df['Cao tầng'] = df['Mô tả'].str.contains(r'cao tầng|lầu|tầng', flags=re.IGNORECASE, na=False).astype(int)
df['Quy hoạch'] = df['Mô tả'].str.contains(r'qui hoạch|quy hoạch', flags=re.IGNORECASE, na=False).astype(int)
df.to_csv("processed_dataset.csv", index=False)
```

```python
df['Giá m2'] = df['Khoảng giá'] / df['Diện tích']
df['Giá m2'] = df['Giá m2'].round(6)
df.to_csv("dataset_cleaned_outliers_new.csv", index=False, decimal=",", sep=";", encoding="utf-8")
```
